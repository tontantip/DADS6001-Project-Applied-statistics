```r
## Morris water maze: number of entries in target area during test trial (3xTg mice)
mice_CTRL <- c(9, 4, 5, 4, 1, 11, 5, 11, 2, 7, 5, 5, 2, 12, 4, 4)
mice_DEF <- c(2, 1, 1, 4, 5, 5, 5, 8, 4, 6, 3, 5, 4, 3, 3, 4, 0, 3, 1, 2, 3, 2)
```

```r
## Number of each data
n_CTRL <- length(mice_CTRL)
n_CTRL # 16
n_DEF <- length(mice_DEF)
n_DEF # 22
```

```r
## Mean and Standard deviation
Mean_CTRL <- mean(mice_CTRL)
Mean_CTRL # 5.6875
Mean_DEF <- mean(mice_DEF)
Mean_DEF # 3.363636
SD_CTRL <- sd(mice_CTRL)
SD_CTRL # 3.380705
SD_DEF <- sd(mice_DEF)
SD_DEF # 1.890967
```

```r
## boxplot
boxplot(mice_CTRL,
        main = "Boxplot CTRL",
        ylab = "Values",
        col = "lightblue",
        border = "blue" )

boxplot(mice_DEF,
        main = "Boxplot DEF",
        ylab = "Values",
        col = "lightblue",
        border = "blue" )
```

```r
## Normality
shapiro.test(mice_CTRL) ## 0.05606
shapiro.test(mice_DEF) ## 0.5769

## Homogeneity of Variances
var.test(mice_CTRL,mice_DEF) ## 0.01489
```

```r
## Two Sample t-test
t.test(mice_DEF,mice_CTRL,var.equal = FALSE , alternative = "less", conf.level = 0.95)
```
